## サンプルデータ1 : 人口データ

```python
from bokeh.sampledata.population import load_population
df = load_population()
```

```python
# 確認
df.head()
```

```python
# 日本に限定
df_Japan = df[df['Location'] == 'Japan']
```

```python
# 年代ごと棒グラフ
# Bar関数を利用
from bkcharts import Bar
p = Bar(df_Japan, 
        label='AgeGrp', 
        values='Value', 
        agg='mean', 
        legend=False, 
        title='年代ごとのValueの平均値')
# show
show(p)
```
